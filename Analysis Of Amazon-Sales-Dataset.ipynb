# Data Cleaning

```python
df.isnull().sum() # found 2 null values in rating_count
```

```python
df[df.isnull().any(axis=1)]
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

### All missing values are now removed 

# Convert actual_price and discount_price into float64

```python
df['discounted_price']=df['discounted_price'].str.replace(r"[‚Çπ,]","",regex=True).astype(float)
```

```python
df['actual_price']=df['actual_price'].str.replace(r"[‚Çπ,]","",regex=True).astype(float)
```

```python
df.info()
```

```python
# Replaces "|" with NaN globally in that column
df['rating'] = df['rating'].astype(str).str.replace("|", "", regex=False)
df['rating'] = df['rating'].str.replace(",", "", regex=False)
```

```python
# 2. Convert to numeric (forces errors to NaN)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df['rating'] = df['rating'].fillna(df['rating'].median())
```

```python
# Verify
print(f"Any nulls left? {df['rating'].isnull().sum()}")
```

```python
df['rating_count']  = df['rating_count'].str.replace("," , "")
```

```python
df[['rating_count','rating']] = df[['rating_count','rating']].astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace("%" , "")
```

```python
df['discount_percentage'] = df['discount_percentage'].astype(float)
```

```python
df.info()
```

```python
# Bivarate Analysis

# Task: Create a Heatmap (sns.heatmap) of the correlation matrix.

# Question to answer: Is there a correlation between discount_percentage and rating_count? (i.e., Do heavily discounted items get more reviews?)

# Create a Correlation Matrix
# Select only the two columns you want to analyze
correlation_matrix = df[['rating', 'discount_percentage']].corr()

# 3. Plot the Heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(data=correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation: Rating vs Discount Percentage')
plt.show()
```

### From the Heatmap above, we see a weak correlation between price and rating. This suggests that expensive products aren't necessarily rated higher by customers.

# 4. Word Cloud (Text Analysis)

Task: You have review_content. Use the wordcloud library to visualize the most common words found in positive reviews vs. negative reviews.



```python
%pip install wordcloud
```

```python
from wordcloud import WordCloud, STOPWORDS

# 1. Prepare the text data
# Join all reviews into a single long string
# We convert values to string just in case there are non-string types
text_data = " ".join(str(review) for review in df['review_content'])

# 2. Define Stopwords (Optional)
# Stopwords are common words like "the", "is", "and" that we want to ignore.
# We can add custom ones specific to your dataset if needed.
custom_stopwords = set(STOPWORDS)
custom_stopwords.update(["Amazon", "product", "bought"]) # Add words you want to remove

# 3. Create the WordCloud object
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white', # 'black' is also common
    stopwords=custom_stopwords,
    min_font_size=10
).generate(text_data)

# 4. Display the generated image using Matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off") # Turn off axis numbers/labels
plt.title("Most Common Words in Amazon Reviews", fontsize=15)
plt.show()
```

## üìù Conclusion & Key Takeaways

In this analysis, we explored the Amazon Sales dataset to understand the relationships between pricing, discounts, and customer satisfaction. After rigorous data cleaning and feature engineering, we derived several key insights:

* **Pricing vs. Satisfaction:** The correlation heatmap revealed a very weak relationship between `Discount Percentage` and `Rating`. This suggests that simply offering a massive discount does not guarantee a 5-star review; product quality remains the primary driver of satisfaction.
* **Customer Sentiment:** The WordCloud analysis highlighted that customers prioritize reliability. Words like **"Quality"**, **"Good"**, **"Working"**, and **"Cable"** were most frequent, indicating that for electronics, functionality is valued over aesthetics.
* **Rating Distribution:** As seen in the Violin Plot, the majority of products in categories like "Computers" and "Electronics" have a high median rating (above 4.0), but there are significant outliers. This indicates a generally positive sentiment on the platform, but with specific low-quality items pulling down the average.
* **Data Quality:** The dataset required significant preprocessing (handling currency symbols, string anomalies like `|`, and type conversion), reinforcing the importance of data cleaning in real-world analysis.

**Final Thought:**
For sellers on Amazon, maintaining product quality appears to be more critical for long-term reputation than aggressive discounting strategies.
